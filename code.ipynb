Conclusion :
Correlation analysis show that most features exhibit weak individual association with fraudulent transactions. The strongest correlation observed was approximately ±0.32, indicating moderate influence. This suggests that fraud detection depends on multi-features rather then indivisual decision making 

```python
corr = df.corr(numeric_only=True)['Class'].drop('Class')
strong_corr = corr[abs(corr) > 0.2].sort_values()
plt.figure(figsize=(8,5))

colors = []

for value in corr:
    if abs(value) > 0.2:        # strong correlation (higher the value more likely to be fraud)
        colors.append('#c3473c')      
    elif abs(value) > 0.1:       
        colors.append('#d8c527')   
    else:                         
        colors.append('#52ad71')    # weak correlation (less likely to be fraud)

strong_corr.plot(kind='bar', color=colors)
plt.title("Strong Feature Correlation with Fraud")
plt.ylabel("Correlation Value")
plt.xlabel("Features")
plt.show()
```
